## Create diagrams that show:
   - c) Deaths per week
   - d) new cases per week
   - e) both deaths and new cases per week
   - f) cumulative cases

```python
fig, ax = plt.subplots(2,2, dpi=500, figsize=(20, 15))
fig.suptitle("Covid-19 statistics in Sweden", fontsize=25)
sns.set_style("whitegrid")


# c) Number of dead per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [0, 0]).set(title="Number of dead per week",
             xlabel="Week", ylabel="Number of dead per week", xticks=covid19["Vecka"][::10])

# d) New cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [0, 1]).set(title="New cases per week",
             xlabel="Week", ylabel="Number of new cases per week", xticks=covid19["Vecka"][::10])

# e) Number of dead and new cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [1, 0])
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [1, 0]).set(title="Number of dead and new cases per week",
             xlabel="Week", ylabel="Numbers per week", xticks=covid19["Vecka"][::10])
ax[1,0].set_yscale("log") # fix scale med log
ax[1,0].legend(["Antal_avlidna_vecka", "Antal_fall_vecka"], labels=["Dead", "New cases"], fontsize=15, loc= "upper left")
# f) Cumulative cases
sns.lineplot(data=covid19, x="Vecka", y="Kum_antal_fall", color="blue",
             ax=ax[1, 1]).set(title="Cumulative cases",
             xlabel="Week", ylabel="Number of cumulative cases per week", xticks=covid19["Vecka"][::10])
# using xticks to decrease frequency based on x: https://stackoverflow.com/questions/12608788/changing-the-tick-frequency-on-x-or-y-axis-in-matplotlib            


plt.savefig("visualiseringar/Covid-19_Statistics_in_Sweden.png")
```

- Use plotly_express to plot the data 


```python
def px_line(dataframe=covid19, x=None, y=None, title=None):
    
    fig = px.line(dataframe, x=x, y=y, markers=True, log_y=True,
        title=title,
        labels=dict(Vecka="Week"))

    fig.update_layout(hovermode="x")
    fig.update_xaxes(
                 showspikes=True,
                 spikedash="solid",
                 spikemode="across",
                 spikecolor="green",
                 spikesnap="cursor"
                 )
    fig.show()
    fig.write_html(f"Visualiseringar/{title}.html") #saving plotly as en html file is based on : https://plotly.com/python/interactive-html-export/
    

px_line(x="Vecka", y="Antal_avlidna_vecka", title="c) Dead per week")
px_line(x="Vecka", y="Antal_fall_vecka", title="d) New cases per week")
px_line(x="Vecka", y=["Antal_avlidna_vecka", "Antal_fall_vecka"], title="e) Dead and new cases per week")
px_line(x="Vecka", y="Kum_antal_fall", title="f) Cumulative cases")
```
