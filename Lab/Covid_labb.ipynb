- b) Slå ihop kolumnerna "år" och "veckonummer" till en kolumn med namn "Vecka"

```python
covid19["år"] = covid19["år"].astype(str) + "v" + covid19["veckonummer"].astype(str)
covid19 = covid19.rename({"år":"Vecka"}, axis=1)
covid19.drop("veckonummer", axis=1)
covid19
```

## Create diagrams that show:
   - c) Deaths per week
   - d) new cases per week
   - e) both deaths and new cases per week
   - f) cumulative cases

```python
fig, ax = plt.subplots(2,2, dpi=500, figsize=(20, 15))
fig.suptitle("Covid-19 statistics in Sweden", fontsize=25)

# c) Number of dead per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [0, 0]).set(title="Number of dead per week",
             xlabel="Week", ylabel="Number of dead per week")

# d) New cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [0, 1]).set(title="New cases per week",
             xlabel="Week", ylabel="Number of new cases per week")

# e) Number of dead and new cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [1, 0])
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [1, 0]).set(title="Number of dead and new cases per week",
             xlabel="Week", ylabel="Numbers per week")


# f) Cumulative cases
sns.lineplot(data=covid19, x="Vecka", y="Kum_antal_fall", color="blue",
             ax=ax[1, 1]).set(title="Cumulative cases",
             xlabel="Week", ylabel="Number of cumulative cases per week")

fig.legend(["Antal_avlidna_vecka", "Antal_fall_vecka"], labels=["Dead", "New cases"], fontsize=15, loc= "center left")

plt.savefig("visualiseringar/Covid-19_Statistics_in_Sweden.png")
```
