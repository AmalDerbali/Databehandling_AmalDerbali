## Create diagrams that show:
   - c) Deaths per week
   - d) new cases per week
   - e) both deaths and new cases per week
   - f) cumulative cases

```python
fig, ax = plt.subplots(2,2, dpi=500, figsize=(20, 15))
fig.suptitle("Covid-19 statistics in Sweden", fontsize=25)
sns.set_style("whitegrid")


# c) Number of dead per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [0, 0]).set(title="Number of dead per week",
             xlabel="Week", ylabel="Number of dead per week", xticks=covid19["Vecka"][::10])

# d) New cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [0, 1]).set(title="New cases per week",
             xlabel="Week", ylabel="Number of new cases per week", xticks=covid19["Vecka"][::10])

# e) Number of dead and new cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [1, 0])
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [1, 0]).set(title="Number of dead and new cases per week",
             xlabel="Week", ylabel="Numbers per week", xticks=covid19["Vecka"][::10])
ax[1,0].set_yscale("log") # fix scale med log
ax[1,0].legend(["Antal_avlidna_vecka", "Antal_fall_vecka"], labels=["Dead", "New cases"], fontsize=15, loc= "upper left")
# f) Cumulative cases
sns.lineplot(data=covid19, x="Vecka", y="Kum_antal_fall", color="blue",
             ax=ax[1, 1]).set(title="Cumulative cases",
             xlabel="Week", ylabel="Number of cumulative cases per week", xticks=covid19["Vecka"][::10])
# using xticks to decrease frequency based on x: https://stackoverflow.com/questions/12608788/changing-the-tick-frequency-on-x-or-y-axis-in-matplotlib            


plt.savefig("visualiseringar/Uppgift1_Covid-19_Statistics_in_Sweden.png")
```

- Use plotly_express to plot the data 


```python
def px_line(dataframe=covid19, x=None, y=None, title=None):
    
    fig = px.line(dataframe, x=x, y=y, markers=True, log_y=True,
        title=title,
        labels=dict(Vecka="Week"))

    fig.update_layout(hovermode="x")
    fig.update_xaxes(
                 showspikes=True,
                 spikedash="solid",
                 spikemode="across",
                 spikecolor="green",
                 spikesnap="cursor"
                 )
    fig.show()
    fig.write_html(f"Visualiseringar/{title}.html") #saving plotly as en html file is based on : https://plotly.com/python/interactive-html-export/
    

px_line(x="Vecka", y="Antal_avlidna_vecka", title="Uppgift1_c) Dead per week")
px_line(x="Vecka", y="Antal_fall_vecka", title="Uppgift1_d) New cases per week")
px_line(x="Vecka", y=["Antal_avlidna_vecka", "Antal_fall_vecka"], title="Uppgift1_e) Dead and new cases per week")
px_line(x="Vecka", y="Kum_antal_fall", title="Uppgift1_f) Cumulative cases")
```

## The number of cases in terms of the number of people vaccinated
- I will start with creating a dataframe and read data:

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly_express as px
import plotly
```

```python
covid_data = pd.read_excel("data/Uppgift_3_Covid19.xlsx", sheet_name="Veckodata Region")
covid_data
```

```python
#Remove region colunm
covid_data = covid_data.drop(labels= "Region", axis = 1)
covid_data
```

```python
#Create a new "Vecka" column and remove år and old week column
covid_data["år"] = covid_data["år"].astype(str) + "v" + covid_data["veckonummer"].astype(str)
covid_data = covid_data.rename({"år":"Vecka"}, axis=1)
covid_data.drop("veckonummer", axis=1)
```

```python
#Create new dataframe related to vaccination statistics
vaccin_data = pd.read_excel("data/Folkhalsomyndigheten_Covid19_Vaccine.xlsx", sheet_name="Vaccinerade tidsserie")
vaccin_data 
```

```python
vaccin_data["Vecka"] = vaccin_data["År"].astype("str")+"v"+ vaccin_data["Vecka"].astype("str")
vaccin_data.head()
```

```python
vaccin_data = vaccin_data.loc[(vaccin_data["Region"] == "| Sverige |")].reset_index(drop=True)
vaccin_data = vaccin_data.drop([0,1])
vaccin_data = vaccin_data.drop(labels= "Region", axis = 1)
vaccin_data
```

```python
#pull the number of people vaccinated with one dose
vaccin_1dos = vaccin_data[vaccin_data["Vaccinationsstatus"]=="Minst 1 dos"].reset_index(drop=True)
vaccin_1dos.head()
```

```python
#pull the number of people who finished the vaccination with 2 doses. 
vaccin_2dos = vaccin_data[vaccin_data["Vaccinationsstatus"]=="Färdigvaccinerade"].reset_index(drop=True)
vaccin_2dos.head()
```

```python
#merging data I need and get it in one column
covid_vaccin = pd.merge(vaccin_2dos, covid_data[["Vecka", "Antal_fall_vecka", "Antal_intensivvårdade_vecka", "Antal_avlidna_vecka"]], 
on="Vecka", how="left")
covid_vaccin
```

```python
#plott figures and save it
plt.figure(dpi=100)
fig = sns.lineplot(data=covid_vaccin, x="Antal vaccinerade", y="Antal_fall_vecka").set(title = "Antal fall mot antal vaccinerade")

plt.savefig("Visualiseringar/Uppgift3_Antal fall mot antal vaccinerade.png")
```

- Interpretation:
The figure shows that the cases per week decreases with the increase of the number of people vaccinated. We can see that there is a slide increase of cases that may indicate different waves of covid-19.

## The number of cases who needed intensive hospital care in terms of the number of peolpe vaccinated. 

```python
plt.figure(dpi=100)
fig = sns.lineplot(data=covid_vaccin, x="Antal vaccinerade", y="Antal_intensivvårdade_vecka").set(
                   title= "Antal intensivvårdade mot antal vaccinerade")
plt.savefig("Visualiseringar/Uppgift3_Antal intensivvårdade mot antal vaccinerade.png")
```

- Interpretation:
This figure shows that there is a decrease in the number of people who needed intensive hospital care in terms of the number of poeple vaccinated. This represent how the vaccination helps at attenuate pressure on hospitals and staff there.

## The number of death in terms of the number of peolpe vaccinated

```python
plt.figure(dpi=100)
fig = sns.lineplot(data=covid_vaccin, x="Antal vaccinerade", y="Antal_avlidna_vecka").set(
                   title= "Antal avlidna mot antal vaccinerade")
plt.savefig("Visualiseringar/Uppgift3_Antal avlidna mot antal vaccinerade.png")
```

- Interpretation:
We can see from the figure that the number of people who died because of the virus decreases and drops rapidly when the number of people vaccinated increases. This may due to the fact that they started to vaccinate older people who are in the risk groups. So, we can conclude from different figures the high efficiency of the vaccin against the spread of Covid19 infection.
