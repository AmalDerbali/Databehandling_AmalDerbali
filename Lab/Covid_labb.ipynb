- b) Slå ihop kolumnerna "år" och "veckonummer" till en kolumn med namn "Vecka"

```python
covid19["år"] = covid19["år"].astype(str) + "v" + covid19["veckonummer"].astype(str)
covid19 = covid19.rename({"år":"Vecka"}, axis=1)
covid19.drop("veckonummer", axis=1)
covid19
```

## Create diagrams that show:
   - c) Deaths per week
   - d) new cases per week
   - e) both deaths and new cases per week
   - f) cumulative cases

```python
fig, ax = plt.subplots(2,2, dpi=500, figsize=(20, 15))
fig.suptitle("Covid-19 statistics in Sweden", fontsize=25)
sns.set_style("whitegrid")


# c) Number of dead per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [0, 0]).set(title="Number of dead per week",
             xlabel="Week", ylabel="Number of dead per week", xticks=covid19["Vecka"][::10])

# d) New cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [0, 1]).set(title="New cases per week",
             xlabel="Week", ylabel="Number of new cases per week", xticks=covid19["Vecka"][::10])

# e) Number of dead and new cases per week
sns.lineplot(data=covid19, x="Vecka", y="Antal_avlidna_vecka", color="red",
             ax=ax [1, 0])
sns.lineplot(data=covid19, x="Vecka", y="Antal_fall_vecka", color="green",
             ax=ax [1, 0]).set(title="Number of dead and new cases per week",
             xlabel="Week", ylabel="Numbers per week", xticks=covid19["Vecka"][::10])

# f) Cumulative cases
sns.lineplot(data=covid19, x="Vecka", y="Kum_antal_fall", color="blue",
             ax=ax[1, 1]).set(title="Cumulative cases",
             xlabel="Week", ylabel="Number of cumulative cases per week", xticks=covid19["Vecka"][::10])
# using xticks to decrease frequency on x: https://stackoverflow.com/questions/12608788/changing-the-tick-frequency-on-x-or-y-axis-in-matplotlib            

ax [1, 0].legend(["Antal_avlidna_vecka", "Antal_fall_vecka"], labels=["Dead", "New cases"], fontsize=15, loc= "upper left")

plt.savefig("visualiseringar/Covid-19_Statistics_in_Sweden.png")
```

- Use plotly_express to plot the data 


```python
def px_line(dataframe=covid19, x=None, y=None, title=None):
    
    fig = px.line(dataframe, x=x, y=y, markers=True, 
        title=title,
        labels=dict(Vecka="Week"))

    fig.update_layout(hovermode="x")
    fig.update_xaxes(
                 showspikes=True,
                 spikedash="solid",
                 spikemode="across",
                 spikecolor="green",
                 spikesnap="cursor"
                 )
    fig.show()
    fig.write_html(f"Visualiseringar/{title}.html") #saving plotly as en html file is base on : https://plotly.com/python/interactive-html-export/

px_line(x="Vecka", y="Antal_avlidna_vecka", title="c) Dead per week")
px_line(x="Vecka", y="Antal_fall_vecka", title="d) New cases per week")
px_line(x="Vecka", y=["Antal_avlidna_vecka", "Antal_fall_vecka"], title="e) Dead and new cases per week")
px_line(x="Vecka", y="Kum_antal_fall", title="f) Cumulative cases")
```

## f) Draw a bar chart with counties in the x-axis and bars for population> 16 years, number at least 1 dose and number fully vaccinated.

```python

from plotly.subplots import make_subplots # obtain 3 bars in the same shart based on :https://plotly.com/python/creating-and-updating-figures/

fig3 = make_subplots()
fig3.add_bar(name="Befolkning", x=vaccin["Län_namn"], y=vaccin["Befolkning"])
fig3.add_bar(name="Antal färdigvaccinerade", x=vaccin["Län_namn"], y=vaccin["Antal färdigvaccinerade"])
fig3.add_bar(name="Antal med minst 1 dos", x=vaccin["Län_namn"], y=vaccin["Antal minst 1 dos"])
      
fig3.update_layout( xaxis_title='Län', yaxis_title='Antal', 
                    title="Befolkning, antal färdigvaccinerade, och antal med minst 1 dos per län")
fig3.update_layout(barmode='group')
fig3.show()
plt.savefig("visualiseringar/Befolkning, antal färdigvaccinerade, och antal med minst 1 dos per län.png")
```
